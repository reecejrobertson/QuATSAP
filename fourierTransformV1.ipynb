# Imports

```python
import numpy as np
from scipy.fftpack import fft, ifft
from matplotlib import pyplot as plt
```

# Constants

```python
DOMAIN = np.arange(61)
plt.rcParams["figure.dpi"] = 300
plt.rcParams["figure.figsize"] = (12,3)
```

# Functions

```python
def plot(domain, frequencies):
    n = len(domain)
    f = lambda domain, frequencies: [np.sum([np.sin(frequency*x) for frequency in frequencies]) for x in domain]
    samples = f(domain, frequencies)
    dft = fft(samples)                          #Otherwise compute the DFT to get the frequencies and magnitudes.
    magnitude = np.abs(dft / n)

    ax1 = plt.subplot(121)                      #Then plot the samples,
    ax1.plot(domain, samples)
    ax1.set_xlabel("Time (Seconds)")
    ax1.set_ylabel("Samples")
    
    ax2 = plt.subplot(122)                      #and the magnitudes.
    ax2.plot(domain[:n//2], magnitude[:n//2])    #Note, we only want to plot the first half of the frequency domain.
    ax2.set_xlabel("Frequency (Hz)")
    ax2.set_ylabel("Magnitude")
    
    plt.suptitle("A Plot of the Sound Wave and the Frequency Domain")
    plt.show()

    print(f'Maximum {"frequency" if len(frequencies) == 1 else "frequencies"}: \t{np.sort(magnitude[:len(magnitude)//2])[-len(frequencies):][::-1]}')
    print(f'Target {"frequency" if len(frequencies) == 1 else "frequencies"}: \t{frequencies}')
```

# Experiment

```python
plot(DOMAIN, [np.pi/6])
```

```python
plot(DOMAIN, [np.pi/12])
```

```python
plot(DOMAIN, [np.pi/6, np.pi/12])
```
